```python
#pie chart
import matplotlib.pyplot as plt
import numpy as np

y = np.array([10,20,30,40])

plt.pie(y)
plt.show() 
```

```python
#with a label parameter
import matplotlib.pyplot as plt
import numpy as np

y = np.array([56, 25, 37, 15])
mylabels = ["vanshika","somi","vaishnavi","shreya"]

plt.pie(y, labels = mylabels)
plt.show() 


```

```python
#pie chart using colors
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["pink","black","blue","yellow"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

```python
#pie chart using legend
import matplotlib.pyplot as plt
import numpy as np

y=np.array([76,86,67,82,90])
x=np.array(["clad","phy","maths","venture","clang"])
mycolors=["orange","pink","red","black","blue"]

plt.pie(y, labels =x, colors= mycolors)
plt.title("marks")
plt.legend(title="subjects")
plt.show() 
```

```python
#using a shadow piechart
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0.8, 0, 0.6]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

```python
#pie chart with a header
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["pink","yellow","green","black"]
mycolor =["yellow","orange","black","green"]

plt.pie(y, labels = mylabels,colors=mycolor)
plt.legend(title = "Four colors:")
plt.show() 
```
